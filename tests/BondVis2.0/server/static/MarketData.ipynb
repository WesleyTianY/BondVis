```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件并筛选特定债券数据
data = pd.read_csv('mkt_price_list_all_20230705.csv')

# print(data['bond_cd'])
filtered_data = data[data['bond_cd'] == 230004]

# print(filtered_data)

# 转换时间戳格式
filtered_data['timeStamp'] = pd.to_datetime(filtered_data['mkt_data_upd_tm'])

# 排序
sorted_data = filtered_data.sort_values(by='timeStamp', ascending=True)

# 创建绘图
plt.figure(figsize=(14, 6))
plt.plot(sorted_data['timeStamp'], sorted_data['wghtd_avg_net_prc'], marker='o', markersize=3)
plt.xlabel('Time')
plt.ylabel('wghtd_avg_net_prc')
plt.title('wghtd_avg_net_prc of Bond 230004 Over Time | Trading sheet')
plt.xticks(rotation=30)
plt.grid(True)

# 调整 y 轴范围
plt.ylim(min(sorted_data['wghtd_avg_net_prc']) - 0.05, max(sorted_data['wghtd_avg_net_prc']) + 0.3)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件并筛选特定债券数据
data = pd.read_csv('mkt_price_list_all_20230705.csv')

# 筛选特定债券数据
filtered_data = data[data['bond_cd'] == 230004]

# 转换时间戳格式
filtered_data['mkt_data_upd_tm'] = pd.to_datetime(filtered_data['mkt_data_upd_tm'])

# 排序
sorted_data = filtered_data.sort_values(by='mkt_data_upd_tm', ascending=True)

# 计算差分
diff_trd_vol = sorted_data['trd_vol'].diff()

# 创建绘图
plt.figure(figsize=(14, 6))
plt.bar(sorted_data['mkt_data_upd_tm'], diff_trd_vol, width=0.003)
plt.xlabel('Time')
plt.ylabel('Difference in trd_vol')
plt.title('Difference in trd_vol of Bond 230004 Over Time | Trading sheet')
plt.xticks(rotation=30)
plt.grid(True)

plt.show()
```
